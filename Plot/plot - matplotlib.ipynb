### 1. 스타일 적용
   - plt.plot(x,y, c=선 색, lw=선두께, ls=선스타일, marker=마커표시, ms=마커크기, mec=마커 선 색, mew=마커선굵기, mfc=마커 내부 색)

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("style")
plt.show()
```

### 2. 축 범위 지정
    - plt.xlim(시작범위, 끝범위) // plt.ylim(시작범위, 끝범위)

```python
plt.title('range')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.ylim(-10,30)
plt.xlim(0,50)
```

### 3. 축 상의 표시 지점 틱 설정
    - plt.xticks() // plt.yticks()
### 4. grid 제거
    - plt.grid(False)

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("tick label")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid(False)
plt.show()
```

### 5. 범례 표시
    - plt.legend(loc=위치 값 표기)
    - 각 그래프 label 설정은 plt.plot()의 옵션으로 삽입한다.

```python
plt.title("one more plot")
plt.plot([1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r", label='one')
plt.plot([9, 16, 4, 1],
         c="k", lw=3, ls=":", marker="s", ms=10, mec="m", mew=5, mfc="c", label='two')
plt.legend(loc=2)
plt.show()
```

### 6. 축 라벨 설정
    - plt.xlabel() // plt.ylabel()

```python
plt.title("one more plot")
plt.plot([1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r", label='one')
plt.plot([9, 16, 4, 1],
         c="k", lw=3, ls=":", marker="s", ms=10, mec="m", mew=5, mfc="c", label='two')
plt.legend(loc=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 7. 서브플롯

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)            # plt.subplot(2,1,1) -> 2행 1열의 플롯 중 첫번째
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()                # tight_layout 명령어는 플롯 간 간격 자동 조정
plt.show()
```

```python
# subplot의 명령어로 복수의 axes 객체를 동시에 생성할 수 있음
# sns.distplot 같은 것은 anova에 있는 방법처럼 해야 하는듯...

fig, axes = plt.subplots(3, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")
axes[2, 0].plot(np.random.rand(5))
axes[2, 0].set_title("axes 5")
axes[2, 1].plot(np.random.rand(5))
axes[2, 1].set_title("axes 6")

plt.tight_layout()
plt.show()
```

```python
# 두개의 그래프를 표기하고 각 y축을 2개로 표시할 때
# twins 명령어로 복수의 y축을 가진 플롯 생성

fig, ax0 = plt.subplots()
ax1 = ax0.twinx()             # ax1과 ax0의 복수 플롯 생성

ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")     # 'r-' : 빨간색의 -선
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.show()
```

## 2. 차트 종류

### (1) bar 차트

```python
import matplotlib as mpl
import matplotlib.pylab as plt

y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

### (1)-2 barh 차트(가로막대)

```python
np.random.seed(0)

people=['몽룡','춘향','방자','향단']
y_pos=np.arange(len(people))
performance=3+10*np.random.rand(len(people))
error=np.random.rand(len(people))

plt.title('barh chart')
plt.barh(y_pos,performance,xerr=error, alpha=0.4)     # alph=0(투명)/=1(불투명)
plt.yticks(y_pos,people)
plt.xlabel('x 라벨')
plt.show()
```

### (2) stem 차트
    - plt.stem() : bar 차트와 유사하지만 폭이 없는 스템플롯 차트
    - linspace(시작값, 끝나는 값, 시작과 끝을 일정한 간격에 따라 x개 생성 De=50개)

```python
x=np.linspace(0.1, 2*np.pi, 10)
plt.title('stem')
plt.stem(x, np.cos(x), '-.')
```

### (3) 파이 차트
    - plt.pie(sizes(크기), labels=파이라벨, explode=띄울거리,autopct=%표기, shodow=True(그림자 설정),startangle=시작 각도 설정 )
    - plt.axis('equal') 설정해줘야 한다는데 뭔지 잘 모르겠음

```python
sizes=[15,30,45,10]
labels=['a','b','c','d']
colors=['yellowgreen','gold','lightskyblue','lightcoral']
explode=(0,0,0.1,0)
plt.title('pie chart')
plt.pie(sizes, labels=labels,explode=explode, colors=colors, autopct='%1.1f%%', shadow=True,startangle=90)
plt.axis('equal')
```

### (4) 히스토그램
    - hist(데이터, bins=(bins 인수로 데이터 집계할 구간 정보를 받음))

```python
np.random.seed(0)
x=np.random.randn(1000)
plt.title('histogram')
arrays, bins, patches=plt.hist(x,bins=10)
```

### (5) 스캐터 플롯(산점도)
    - plt.scatter()

```python
np.random.seed(0)
x=np.random.normal(0,1,100)
y=np.random.normal(0,1,100)
plt.title('scatter plot')
plt.scatter(x,y)
```

### (6) 버블차트 : 스캐터 차트 중 데이터가 2차원이 아닌 3,4 차원인 경우
    - plt.scatter(x, y, c=인수 , s=크기 )

```python
n=30
np.random.seed(0)
x=np.random.rand(n)
y1=np.random.rand(n)
y2=np.random.rand(n)
y3=np.pi*(15*np.random.rand(n))**2
plt.title('bubble chart')
plt.scatter(x,y1,c=y2,s=y3)
```
