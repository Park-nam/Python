```python
# seaborn은 matplotlib을 기반으로 다양한 색상 테마와 통계용 차트 등의 기능을 추가한 시각화 패키지

# 샘플데이터 로드
%matplotlib inline

iris=sns.load_dataset('iris')          # 붓꽃 데이터
titanic = sns.load_dataset('titanic')  # 타이타닉호 데이터
tips = sns.load_dataset('tips')        # 팁 데이터
flights = sns.load_dataset('flights')  # 여객운송 데이터
```

## 1. 1차원 데이터

### (1) rug 플롯 : 데이터 위치를 x축 위에 작은 선분(rug)로 나타내어 데이터 위치 표기

```python
x=iris.petal_length.values

sns.rugplot(x)
plt.title('rugplot')
plt.show()
```

### (2) 커널밀도(kernel density) : 커널 함수를 겹치는 방법으로 히스토그램보다 부드러운 형태의 분포 곡선

```python
sns.kdeplot(x)
plt.title('kernel desity, kedplot')
```

### (3) distplot : 러그와 커널 밀도 표시 기능이 있는 히스토그램 플롯 생성
    - default 값으로 히스토그램 생성 후 ked=True와 rug=True 옵션으로 실행

```python
sns.distplot(x, kde=True, rug=True)
```

### (4) 카운트플롯 : countplot 명령을 사용하면 각 카테고리 값별로 데이터가 얼마나 있는지 표시
    - 카운트플롯은 데이터프레임에서만 사용 가능
    -  countplot(x='분석할 열의 이름', data=분석할 데이터의 이름)

```python
sns.countplot(x='pclass', data=titanic)
```

### (4)-1 histogram과 countplot의 비교

```python
plt.hist(titanic.pclass.values)
```

## 2. 2차원 데이터

### (1) jointplot
    - jointplot 명령어의 차트종류를 사용하여 플롯 생성(가장자리에 히스토그램 출력)
    - jointplot(x=사용할 x 변수, y=사용할 y변수, data=사용할 데이터프레임, kind=차트의 종류)

```python
sns.jointplot(x='sepal_length',y='sepal_width', data=iris, kind='scatter') # kind의 default는 scatter인듯
plt.suptitle('jointplot', y=1.02)
```

### (1)-2 kind='kde'면 커널 밀도 히스토그램 생성

```python
sns.jointplot(x=iris.sepal_length, y=iris.sepal_width, data=iris, kind='kde')
```

### (2) barplot : 카테고리 값에 따른 실수 값의 평균과 편차를 표시
     - 평균은 막대의 높이로, 편차는 에러바(error bar)로 표시

```python
sns.barplot(x=tips.day, y=tips.total_bill, data=tips)
```

### (2)-2 hue 옵션을 사용한 barplot

```python
sns.barplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex)
```

### (3) boxplot

```python
sns.boxplot(x=tips.day, y=tips.total_bill, data=tips)
```

### (3)-2 hue 옵션을 사용한 boxplot

```python
sns.boxplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex)
```

### (4) violinplot
     - boxplot은 중앙값, 표준편차 등 분포의 간략한 특성을 보여주는데 반해
     - violinplot, stripplot, swarmplot 등은 카테고리값에 따른 각 분포의 실제 데이터나 전체 형상을 보여줌

```python
sns.violinplot(x=tips.day, y=tips.total_bill, data=iris)
```

### (4)-2 hue 옵션을 사용한 violinplot

```python
sns.violinplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex)
```

### (4)-3 split 옵션(시각화 방법 변경, stripplot, viloinplot, swarmplot 사용 가능)

```python
sns.violinplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex, split=True)
```

### (4)-4 dodge 옵션 사용

```python
sns.violinplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex, dodge=True)
```

### (5) stripplot
    - scatter plot과 비슷한 형상으로 데이터를 점으로 표현
    - jitter=True 옵션을 설정하면 가로축 상의 위치를 무작위로 바꾸어 데이터 수가 많을 경우 겹치지 않도록 함

```python
sns.stripplot(x=tips.day, y=tips.total_bill, data=tips)
```

```python
np.random.seed(0)
sns.stripplot(x=tips.day, y=tips.total_bill, data=tips, jitter=True)

# jitter의 역할이 무엇인지 잘 모르겠음...
```

### (5)-2 hue 옵션을 사용한 stripplot

```python
sns.stripplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex)
plt.legend(loc=1)
```

### (5)-3 split 옵션 적용한 stripplot

```python
sns.stripplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex, split=True)
```

### (5)-4 dodge 옵션 적용한 stripplot

```python
sns.stripplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex, dodge=True)
```

### (6) swarmplot
    - stripplot과 유사하나 데이터를 나타내는 점이 겹치지 않도록 출력

```python
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips)
```

### (6)-2 hue 옵션을 사용한 swarmplot

```python
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex)
plt.legend(loc=1)
```

### (6)-3 split 옵션을 사용한 swarmplot

```python
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex, split=True)
```

### (6)-3 dodge 옵션을 사용한 swarmplot

```python
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex, dodge=True)
```

## 3. 다차원 실수형 데이터(pairplot 명령어 사용)
    - pairplot명령어를 사용해 각 열의 조합에 대해 scatter 플롯 생성

```python
sns.pairplot(iris)
```

* 만약 카테고리형 데이터가 섞여 있는 경우 hue 옵션에 카테고리 변수 이름 지정하여 카테고리 값에 따라 색상을 다르게 할 수 있다.

```python
sns.pairplot(iris, hue='species',markers=['o','s','D'])
```

```python
# 각 그래프는 곂쳐서 사용할 수 있다.

sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)
sns.stripplot(x="tip", y="day", data=tips, jitter=True, color="0.4")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips, color="0.9")
plt.show()
```
