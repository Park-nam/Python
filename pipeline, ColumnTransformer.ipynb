```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
house = pd.read_csv('C:/Users/Park/_Python_data/houseprice/train.csv')
house.drop(columns = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'Id','FireplaceQu'], inplace = True, axis = 1)

X = house.drop(['SalePrice'],axis=1)
y = house.SalePrice
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

# 파이프라인을 사용하지 않은 전처리

- 수치형 - 스케일링, 결측값
- 카테고리형 - 인코딩, 결측값

```python
num_col = []
num = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
cat_col = []
cat = ['object']

for i in X_train.columns:
    if X_train[i].dtype in num:
        num_col.append(i)
        
for i in X_train.columns:
    if X_train[i].dtype in cat:
        cat_col.append(i)
        
X_train_num = X_train[num_col]
X_train_cat = X_train[cat_col]

X_train_cat.fillna('NaN', axis=1, inplace = True)
X_train_num.fillna(X_train_num.mean(),inplace=True)

one_hot = OneHotEncoder(sparse = False)
encoder = one_hot.fit_transform(X_train_cat)
X_train_cat_enco = pd.DataFrame(encoder, columns = one_hot.get_feature_names(cat_col))

scaler = StandardScaler()
scale=scaler.fit_transform(X_train_num)
X_train_num_scale=pd.DataFrame(scale, columns=num_col)

X_train_df = pd.concat([X_train_num_scale,X_train_cat_enco],axis=1)

X_train_df
```

# 파이프라인을 이용한 전처리

- 수치형 - 스케일링, 결측값
- 카테고리형 - 인코딩, 결측값
- 데이터를 분할하기 전에 실시하는 것이 좋다
- 만약 데이터가 분할되기 전이라면, X_train에 fit_transform 후 나머지에 transform만 적용

### 1) 학습/평가 데이터 분할 전

```python
# columntransformer를 위한 컬럼 리스트 생성
numerical = X.select_dtypes(include='number').columns
categorical = X.select_dtypes(include='object').columns              

# numeric, categoric 파이프라인 생성
num_pipeline = Pipeline([('imputer',SimpleImputer(strategy = 'mean')),   
                        ('scaler',StandardScaler())])
# constant는 특정값으로 대체한다는 의미로 missing 데이터로 대체
# handle_unknown 파라미터를 사용하는 이유는 학습 및 검증 데이터를 분할하면서 서로에게 포함되지 않는 변수가 발생할 수 있기 때문
cat_pipeline = Pipeline([('imputer',SimpleImputer(strategy = 'constant', fill_value='None')),
                         ('encoder',OneHotEncoder(sparse = False, handle_unknown = 'ignore'))])

# columntransformer 생성
preprocessor = ColumnTransformer([('num', num_pipeline, numerical),('cat', cat_pipeline, categorical)])

# columntransformer로 데이터 학습
X = preprocessor.fit_transform(X)

# categorical 데이터의 컬럼명 추출 후 nemerical 컬럼명과 병합
cat_columns = preprocessor.named_transformers_['cat']['encoder'].get_feature_names(categorical)
columns = np.append(numerical, cat_columns)

# 데이터프레임 결합
X = pd.DataFrame(X, columns = columns)

display(X)
```

### 2) 학습/평가 데이터 분할 후

```python
# columntransformer를 위한 컬럼 리스트 생성
numerical = X_train.select_dtypes(include='number').columns
categorical = X_train.select_dtypes(include='object').columns              

# numeric, categoric 파이프라인 생성
num_pipeline = Pipeline([('imputer',SimpleImputer(strategy = 'mean')),   
                        ('scaler',StandardScaler())])
# constant는 특정값으로 대체한다는 의미로 missing 데이터로 대체
# handle_unknown 파라미터를 사용하는 이유는 학습 및 검증 데이터를 분할하면서 서로에게 포함되지 않는 변수가 발생할 수 있기 때문
cat_pipeline = Pipeline([('imputer',SimpleImputer(strategy = 'constant', fill_value='missing')),
                         ('encoder',OneHotEncoder(sparse = False,handle_unknown = 'ignore'))])

# columntransformer 생성
preprocessor = ColumnTransformer([('num', num_pipeline, numerical),('cat', cat_pipeline, categorical)])

# columntransformer로 데이터 학습
fit = preprocessor.fit(X_train)

# categorical 데이터의 컬럼명 추출 후 nemerical 컬럼명과 병합
cat_columns = preprocessor.named_transformers_['cat']['encoder'].get_feature_names(categorical)
columns = np.append(numerical, cat_columns)

# 데이터프레임 결합
X_train = pd.DataFrame(fit.transform(X_train), columns = columns)
X_test = pd.DataFrame(fit.transform(X_test), columns = columns)

display(X_train)
display(X_test)
```
