## 1. 홈페이지 데이터 추출

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://comic.naver.com/webtoon/weekday'
res = requests.get(url)
res.status_code
soup = BeautifulSoup(res.text, 'lxml')
```

## 2-1 출력하고자 하는 데이터 값 선택

```python
# f12를 눌렀을 때 title 출력
print(soup.title)

print('-'*30)
# 타이틀의 텍스트만 출력
print(soup.title.get_text())

print('-'*30)
# div에서 처음 발견되는(첫번째의) a element를 출력
print(soup.a)

print('-'*30)
# a element의 속성 정보 출력
print(soup.a.attrs)

print('-'*30)
# a element의 herf의 속성값 출력
print(soup.a['href'])
```

## 2-2 find 함수를 통해 조건값 설정

- 위의 방법은 홈페이지의 구조를 잘 알 때 사용가능하다
- 하지만 일반적으로 홈페이지의 구조를 잘 알기 어렵다
- 따라서 find 함수를 통해 조건을 줄 수 있다.

```python
# class 속성이 Nbtn_upload인 a element 출력
print(soup.find('a', attrs = {'class':'Nbtn_upload'})) # class = Nbtn_upload 인 a element 출력
print('-'*30)
print(soup.find(attrs = {'class':'Nbtn_upload'}))      # class = Nbtn_upload 인 어떤 element 출력
                                                       #(웹툰 올리기 버튼이 1개만 존재하여 사실상 동일한 값이 출력된다고 볼 수 있음)
    
# class 속성이 rank01인 li element 반환
rank1 = soup.find('li', attrs = {'class':"rank01"})
print('-'*30)
print(rank1.a)

# text 속성이 프리드로우인 a element 반환
print('-'*30)
webtoon = soup.find('a', text = '프리드로우')
print(webtoon)
```

## 2-3 부모, 형제 관계 간의 element 출력

- next_sibling : 형제 관계 다음 element로 이동
- previous_sibling : 형제 관계 이전 element로 이동
- parent : 부모 관계 element로 이동

```python
# next_sibling을 두번 해주는 이유는 줄바꿈 등이 있을 수 있기 때문
rank2 = rank1.next_sibling.next_sibling
rank3 = rank2.next_sibling.next_sibling


print('rank 1 :', rank1.a.get_text())

print('-'*30)
print('rank 2 :', rank2.a.get_text())

print('-'*30)
print('rank 3 :', rank3.a.get_text())

print('-'*30)
# rank1의 부모 element로 전체 순위관련 요소를 출력한다
print(rank1.parent)
```

## 2-4 find 함수로 조건값을 통해 element 출력

- find 메서드를 사용하여 조건을 추가할 수 있다.
- find_next_sibling('조건값')
- find_previous_sibling('조건값')
- find_parent('조건값')

```python
# rank1의 형제관계 다음 element 중 li 를 출력
rank2 = rank1.find_next_sibling('li')
print(rank2.a.get_text())
print('-'*30)

rank3 = rank2.find_next_sibling('li')
print(rank3.a.get_text())
```

## 2-5 find_next_siblings로 모든 요소 출력

- 일괄적으로 모든 정보를 출력하는 방법
- find_next_siblings('조건값') : 다음 형제 element를 모두 가져옴
- find_previous_siblings('조건값') : 이전 형제 element를 모두 가져옴
- ...

```python
rank1.find_next_siblings('li')
```

# 3. find_all을 이용한 모든 데이터 출력

- find : 첫번째 값 출력
- find_all : 만족하는 모든 값 출력

```python
# 웹툰 페이지에서 월별 웹툰의 작은 제목을 보면 class는 title이며, 태그명이 a인 모든 항목 출력

# class 속성이 title인 모든 a element 반환
cartoons=soup.find_all('a', attrs={'class':'title'})

for cartoons in cartoons:
    print(cartoons.get_text())
```

## 3-1 find_all을 이용하여 특정 웹툰 회차별 제목 출력

```python
url = 'https://comic.naver.com/webtoon/list?titleId=183559&weekday=mon'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml')
res.status_code
```

```python
tower_of_god = soup.find_all("td", attrs={"class":"title"})

print('첫번째 제목 :', tower_of_god[0].a.get_text())
# 한가지 주의(?)할 점은 herf 속성값의 경우 ~~naver.com 이후 데이터만 존재한다. 따라서 앞의 ~~naver.com을 따로 붙여주어야 한다.
print('첫번째 링크 :', 'https://comic.naver.com'+tower_of_god[0].a['href'])
```

```python
for i in tower_of_god:
    title = i.a.get_text()
    link = 'https://comic.naver.com' + i.a['href']
    print(title,link)   
    
```
